```python
import pandas as pd
total = pd.read_csv("./drive/MyDrive/733_project/weather_merged/etl_filtered.csv")
total=total.iloc[:,1:].drop(["TOBS","SNOW","SNWD"],axis=1)

total["datetime"]=total.datetime.astype('datetime64[ns]')
```

```python
import datetime
def datetime_to_float(d):
    dtime=d["datetime"]
    epoch = datetime.datetime.utcfromtimestamp(0)
    total_seconds =  (dtime - epoch).total_seconds()
    # total_seconds will be in decimals (millisecond precision)
    return total_seconds
total["second"]=total.apply(datetime_to_float,axis=1)
```

```python

# copy the data
max_scaled = total.copy()

for column in ["second","LONGITUDE_x","LATITUDE_x"]:
    max_scaled[column] = max_scaled[column]  / max_scaled[column].abs().max()
max_scaled
```

```python
unfilled_table=max_scaled[max_scaled["ELEVATION"].isna()]
unfilled_table
filled_table=max_scaled[max_scaled["ELEVATION"].notnull()]
filled_table
```

```python
from sklearn.linear_model import LinearRegression

unfilled_X=unfilled_table[["second","LATITUDE_x","LONGITUDE_x"]]
unfilled_y = unfilled_table[["ELEVATION"]]
filled_X=filled_table[["second","LATITUDE_x","LONGITUDE_x"]]
filled_y=filled_table[["ELEVATION"]]

model_unfilled=LinearRegression()
model_filled=LinearRegression()
# lr_unfilled=model_unfilled.fit(unfill_X,PSID_y)
lr_filled=model_filled.fit(filled_X,filled_y)

unfilled_table["ELEVATION"]=lr_filled.predict(unfilled_X)
max_scaled.loc[unfilled_table.index,"ELEVATION"]=unfilled_table["ELEVATION"]
max_scaled
```

```python
total=total.drop(["second"],axis=1)
total["TMIN"]=max_scaled["TMIN"]
total["TMAX"]=max_scaled["TMAX"]
total["PRCP"]=max_scaled["PRCP"]
total["AWND"]=max_scaled["AWND"]
total["ELEVATION"]=max_scaled["ELEVATION"]
total.to_csv('./drive/MyDrive/733_project/weather_merged/etl_filtered_v1.csv')
```
