```python
# encoding state
enc = OneHotEncoder(handle_unknown='ignore')
X = df['STATE'].to_numpy().reshape(-1, 1)
enc.fit(X)
ec_df = pd.DataFrame(enc.transform(X).toarray())
ec_df = df.join(ec_df)
ec_df
```

```python
df.columns
```

```python
mean_pd = df[df['population_density']>=0].groupby('STATE')['population_density'].mean().reset_index()
# print(mean_pd)
ml_df = df[['month', 'STATE', 'TMAX', 'TMIN']]
ml_df = ml_df.groupby(['month', 'STATE']).agg({'month' : ['count'], 'TMAX' : 'mean', 'TMIN' : 'mean'}).reset_index()
# ml_df.columns = ml_df.columns.get_level_values(0)
ml_df.columns = [' '.join(col).strip() for col in ml_df.columns.values]
# print(ml_df)
merged = ml_df.merge(mean_pd, how='inner', left_on='STATE', right_on='STATE')
encoding = pd.get_dummies(merged.STATE)
merged = merged.join(encoding)
merged
```

```python
import autosklearn.regression 
```

```python
y = merged['month count'].to_frame()
X = merged.drop(['month count', 'STATE'], axis=1)
automl = autosklearn.regression.AutoSklearnRegressor(
    time_left_for_this_task=120,
    per_run_time_limit=30,
    tmp_folder='/content/tmp1'
)
automl.fit(X, y)
```

```python
import sklearn
train_predictions = automl.predict(X)
print("Train R2 score:", sklearn.metrics.r2_score(y, train_predictions))
```
