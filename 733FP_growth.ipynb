```python
from pyspark.ml.fpm import FPGrowth

df = spark.createDataFrame([
    (0, [1, '2', 5]),
    (1, [1, '2', 3, 5]),
    (2, [1, '2'])
], ["id", "items"])

df.show()

fpGrowth = FPGrowth(itemsCol="items", minSupport=0.5, minConfidence=0.6)
model = fpGrowth.fit(df)

# Display frequent itemsets.
model.freqItemsets.show()

# Display generated association rules.
model.associationRules.show()

# transform examines the input items against all the association rules and summarize the
# consequents as prediction
model.transform(df).show()
```

```python
import pandas as pd
from dataprep.eda import plot
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/733project/AL.csv')
features = ['FIRE_YEAR', 'STAT_CAUSE_CODE', 'STAT_CAUSE_DESCR', 'FIRE_SIZE_CLASS', 
       'STATE', 'population_density', 'ELEVATION', 'DATE', 'TMAX', 'TMIN']
fp_df = df[features]
fp_df
```

```python
plot(fp_df)
```

```python
fp_df['STAT_CAUSE_CODE'] = fp_df['STAT_CAUSE_CODE'].apply(lambda x: 'stat_cause_code_'+str(int(x)))
```

```python
fp_df['FIRE_SIZE_CLASS'] = fp_df['FIRE_SIZE_CLASS'].apply(lambda x: 'fire_size_class_'+x)
```

```python
fp_df['population_density'] = fp_df['population_density'].apply(lambda x: 'population_density_'+str((int(x/10))))
```

```python
fp_df['ELEVATION'] = fp_df['ELEVATION'].apply(lambda x: 'elevation_'+str(int(x/10)) if x>=0 else 'elevation_-1')
```

```python
fp_df['year'] = fp_df['DATE'].apply(lambda x: 'year'+str(x[0:4]))
fp_df['month'] = fp_df['DATE'].apply(lambda x: 'month'+str(x[5:7]))
fp_df['TMAX'] = fp_df['TMAX'].apply(lambda x: 'TMAX_'+str(int(x/10)) if x>=0 else 'TMAX_-1')
fp_df['TMIN'] = fp_df['TMIN'].apply(lambda x: 'TMIN_'+str(int(x/10)) if x>=0 else 'TMIN_-1')
fp_df
```

```python
from pyspark.sql.functions import split
spark_df = (spark.read
    .text("/content/fp_item.txt")
    .select(split("value", ",").alias("items")))
spark_df.show()
```

```python
fp = FPGrowth(minSupport=0.3, minConfidence=0.3)
model = fp.fit(spark_df)
model.freqItemsets.show()

# Display generated association rules.
model.associationRules.show()

# transform examines the input items against all the association rules and summarize the
# consequents as prediction
model.transform(spark_df).show()
```

```python
a = model.freqItemsets
a.show()
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def array_to_string(my_list):
    return '[' + ','.join([str(elem) for elem in my_list]) + ']'
array_to_string_udf = udf(array_to_string, StringType())
a = a.withColumn('items', array_to_string_udf(a['items']))
a.write.csv('output')
```
